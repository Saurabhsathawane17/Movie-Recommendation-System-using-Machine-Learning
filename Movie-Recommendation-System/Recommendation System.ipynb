```python
# now we have to convert 
#list of dictionary to list
# for e.g:-["Action","Adventure","Fantasy","Science Fiction"]
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"].apply(convert)
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
# now we want original actor "name" from list of cast
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter +=1
        else:
           break
        return L
```

```python
movies["cast"].apply(convert3)
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
# now we wan director name from "crew" column
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
        return l
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```
